```python
#Importando o pacote pandas
import pandas as pd
```

```python
# Carregando os dados e armazenando em um Dataframe
df_notas = pd.read_excel('notas.xlsx')
```

```python
# Visualizando o Dataframe
df_notas
```

```python
# Renomeia as variaves
df_notas.rename(columns = {'ID':'ID_ALUNO', 'DISCIPLINA':'MATÉRIA', 'NT':'NOTAS'}, inplace = True)
```

```python
# Vamos descartar os valores duplicados
df_notas.drop_duplicates(['ID_ALUNO', 'MATÉRIA', 'NOTAS'], inplace=True)
```

```python
# Visualiando a mediana dos dados
df_notas['NOTAS'].median()
```

```python
# Preenchendo os valores NaN com a mediana dos dados
df_notas['NOTAS'] = df_notas['NOTAS'].fillna((df_notas['NOTAS'].median()))
```

```python

# Preenchendo valores nulos com a média
df_notas = pd.read_excel('notas.xlsx')
df_notas.rename(columns = {'ID':'ID_ALUNO', 'DISCIPLINA':'MATÉRIA', 'NT':'NOTAS'}, inplace = True)
df_notas.drop_duplicates(['ID_ALUNO', 'MATÉRIA', 'NOTAS'], inplace=True)
df_notas
```

```python
# Preenchendo os valores NaN com a media dos dados
df_notas['NOTAS'] = df_notas['NOTAS'].fillna((df_notas['NOTAS'].mean()))
df_notas
```
